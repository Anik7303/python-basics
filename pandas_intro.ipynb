# label based data extraction:
single value: use loc[index_label,column_label]

range: use loc[start_index_label: end_index_label, start_column_label: end_colum_label]

```python
import pandas
df = pandas.read_csv('data.csv')
df1 = df.set_index('id')
df1.loc[2,'name']
df1.loc[2:5, 'city':'name']
```

# index based extraction:

single value: use `iloc[index_number,column_number]`

range: use `iloc[start_index_number: end_index_number, start_column_number: end_colum_number]`

```python
import pandas
df = pandas.read_csv('data.csv')
df1 = df.set_index('id')
df1.iloc[1,4]
df1.iloc[2:5,1:4] # 2 to 4, 1 to 3
```

# drop row & columns:

## drop a row:
```python
import pandas
df = pandas.read_csv('data.csv')
df1 = df.set_index('id')
df2 = df1.drop(2, 0) # 0 for rows
df3 = df1.drop(df1.index[2:4], 0) # range delete
```

## drop a column:
```python
import pandas
df = pandas.read_csv('data.csv')
df2 = df1.drop('city', 1) # 1 for columns
df3 = df1.drop(df1.columns[1:3], 1) # range delete
```
